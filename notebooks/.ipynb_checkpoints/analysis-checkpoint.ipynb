# Análisis Cuantitativo de SBOMs y Vulnerabilidades

Este notebook analiza los resultados de la generación de SBOMs (con **Syft**) y el análisis de vulnerabilidades (con **Grype**) sobre los repositorios públicos activos de una organización de GitHub.

## Estructura del proyecto
```
sbom-project/
├── scripts/
│   ├── 01_fetch_repos.py          # Recupera repos activos de GitHub
│   ├── 02_generate_sboms.py       # Genera SBOMs con Syft
│   └── 03_analyze_vulnerabilities.py  # Analiza con Grype
├── sboms/                         # Archivos .cdx.json por repo
├── reports/
│   ├── repos.json
│   ├── sbom_summary.json
│   ├── vulnerability_report.json
│   └── grype_raw/                 # Salida cruda de Grype por repo
└── notebooks/
    └── analysis.ipynb             # ← Este notebook
```

## Reproducibilidad
Para reproducir el análisis completo:
```bash
# 1. Instalar dependencias
pip install -r requirements.txt

# 2. Instalar herramientas externas
# Syft: https://github.com/anchore/syft
curl -sSfL https://raw.githubusercontent.com/anchore/syft/main/install.sh | sh -s -- -b /usr/local/bin
# Grype: https://github.com/anchore/grype
curl -sSfL https://raw.githubusercontent.com/anchore/grype/main/install.sh | sh -s -- -b /usr/local/bin

# 3. Ejecutar pipeline
export GITHUB_TOKEN=<tu_token>
python scripts/01_fetch_repos.py <ORGANIZACIÓN> --max-repos 50
python scripts/02_generate_sboms.py
python scripts/03_analyze_vulnerabilities.py

# 4. Abrir notebook
jupyter notebook notebooks/analysis.ipynb
```

````python
# ── Dependencias ────────────────────────────────────────────────────────────
import json
import os
from pathlib import Path
from collections import Counter, defaultdict

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from IPython.display import display, Markdown

# Estilo
sns.set_theme(style='whitegrid', palette='muted', font_scale=1.1)
plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (12, 5)

# Rutas (relativas al notebook)
ROOT = Path('..') 
REPORTS = ROOT / 'reports'
SBOMS   = ROOT / 'sboms'

print('Rutas configuradas:')
for p in [REPORTS, SBOMS]:
    print(f'  {p.resolve()} — existe: {p.exists()}')
````

---
## 1. Carga de datos

````python
# ── Cargar repos.json ────────────────────────────────────────────────────────
with open(REPORTS / 'repos.json') as f:
    repos_raw = json.load(f)

ORG = repos_raw['org']
df_repos = pd.DataFrame(repos_raw['repos'])
df_repos['pushed_at'] = pd.to_datetime(df_repos['pushed_at'])

display(Markdown(f'### Organización: `{ORG}`'))
display(Markdown(f'**Repositorios activos encontrados:** {len(df_repos)}'))
df_repos[['name','language','size_kb','stargazers_count','pushed_at']].head(10)
````

````python
# ── Cargar vulnerability_report.json ─────────────────────────────────────────
with open(REPORTS / 'vulnerability_report.json') as f:
    vuln_raw = json.load(f)

results = [r for r in vuln_raw['results'] if 'error' not in r]
print(f'Repos analizados con éxito: {len(results)}/{vuln_raw["total_repos"]}')

# DataFrame de resumen por repo
rows = []
for r in results:
    sc = r.get('severity_counts', {})
    rows.append({
        'repo': r['repo'],
        'total_vulns': r['total_vulnerabilities'],
        'critical': sc.get('Critical', 0),
        'high':     sc.get('High', 0),
        'medium':   sc.get('Medium', 0),
        'low':      sc.get('Low', 0),
        'negligible': sc.get('Negligible', 0),
        'unknown':  sc.get('Unknown', 0),
        'packages_affected': r.get('packages_affected_count', 0),
        'max_cvss': r.get('max_cvss_score', 0.0),
        'fixable':  r.get('fixable', 0),
        'analyzed_at': r.get('analyzed_at'),
    })

df = pd.DataFrame(rows).sort_values('total_vulns', ascending=False).reset_index(drop=True)
df.head()
````

````python
# ── Cargar sbom_summary.json ──────────────────────────────────────────────────
with open(REPORTS / 'sbom_summary.json') as f:
    sbom_raw = json.load(f)

sbom_rows = []
for s in sbom_raw['sboms']:
    if 'error' in s:
        continue
    for eco, count in s.get('ecosystems', {}).items():
        sbom_rows.append({'repo': s['repo'], 'ecosystem': eco, 'components': count})

df_sbom = pd.DataFrame(sbom_rows) if sbom_rows else pd.DataFrame(columns=['repo','ecosystem','components'])

df_sbom_summary = pd.DataFrame([
    {'repo': s['repo'], 'total_components': s.get('component_count', 0)}
    for s in sbom_raw['sboms'] if 'error' not in s
])

print(f'Repos con SBOM: {len(df_sbom_summary)}')
print(f'Total componentes: {df_sbom_summary["total_components"].sum():,}')
````

---
## 2. Análisis de Repositorios

````python
# ── Lenguajes de programación ────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribución de lenguajes
lang_counts = df_repos['language'].fillna('Unknown').value_counts().head(12)
axes[0].barh(lang_counts.index[::-1], lang_counts.values[::-1], color=sns.color_palette('muted', len(lang_counts)))
axes[0].set_title('Lenguajes de programación\n(top 12 repos activos)', fontweight='bold')
axes[0].set_xlabel('Número de repositorios')
for i, v in enumerate(lang_counts.values[::-1]):
    axes[0].text(v + 0.1, i, str(v), va='center')

# Tamaño de repos
axes[1].hist(df_repos['size_kb'].clip(upper=df_repos['size_kb'].quantile(0.95)),
             bins=20, color='steelblue', edgecolor='white')
axes[1].set_title('Distribución de tamaño de repositorios\n(percentil 95)', fontweight='bold')
axes[1].set_xlabel('Tamaño (KB)')
axes[1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.savefig(REPORTS / 'fig_repos.png', bbox_inches='tight')
plt.show()
````

---
## 3. Análisis de Dependencias (SBOMs)

````python
# ── Ecosistemas / gestores de paquetes ──────────────────────────────────────
eco_total = df_sbom.groupby('ecosystem')['components'].sum().sort_values(ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Barras de ecosistemas
palette = sns.color_palette('tab10', len(eco_total))
axes[0].bar(eco_total.index, eco_total.values, color=palette)
axes[0].set_title('Total de componentes por ecosistema', fontweight='bold')
axes[0].set_xlabel('Ecosistema')
axes[0].set_ylabel('Número de componentes')
axes[0].tick_params(axis='x', rotation=45)
for i, v in enumerate(eco_total.values):
    axes[0].text(i, v + eco_total.max()*0.01, f'{v:,}', ha='center', fontsize=8)

# Pie chart
top_n = min(8, len(eco_total))
pie_data = eco_total.head(top_n)
if len(eco_total) > top_n:
    pie_data['otros'] = eco_total.iloc[top_n:].sum()
axes[1].pie(pie_data, labels=pie_data.index, autopct='%1.1f%%',
            colors=sns.color_palette('pastel', len(pie_data)), startangle=140)
axes[1].set_title('Proporción de componentes\npor ecosistema', fontweight='bold')

plt.tight_layout()
plt.savefig(REPORTS / 'fig_ecosystems.png', bbox_inches='tight')
plt.show()
````

````python
# ── Componentes por repo (top 20) ────────────────────────────────────────────
top_components = df_sbom_summary.sort_values('total_components', ascending=False).head(20)

plt.figure(figsize=(13, 6))
colors = sns.color_palette('Blues_d', len(top_components))
bars = plt.barh(top_components['repo'][::-1], top_components['total_components'][::-1], color=colors)
plt.title('Top 20 repositorios por cantidad de componentes en SBOM', fontweight='bold')
plt.xlabel('Componentes detectados')
for bar in bars:
    w = bar.get_width()
    plt.text(w + top_components['total_components'].max()*0.005, bar.get_y() + bar.get_height()/2,
             f'{int(w):,}', va='center', fontsize=8)
plt.tight_layout()
plt.savefig(REPORTS / 'fig_components_per_repo.png', bbox_inches='tight')
plt.show()

print(f'\nEstadísticas de componentes por repo:')
print(df_sbom_summary['total_components'].describe().apply(lambda x: f'{x:,.1f}'))
````

---
## 4. Análisis de Vulnerabilidades

````python
# ── Resumen global ────────────────────────────────────────────────────────────
total_vulns   = df['total_vulns'].sum()
total_critical = df['critical'].sum()
total_high    = df['high'].sum()
total_medium  = df['medium'].sum()
total_low     = df['low'].sum()
total_fixable = df['fixable'].sum()

summary_md = f"""
| Métrica | Valor |
|---|---|
| Repos analizados | {len(df)} |
| Total vulnerabilidades | {total_vulns:,} |
| 🔴 Critical | {total_critical:,} |
| 🟠 High | {total_high:,} |
| 🟡 Medium | {total_medium:,} |
| 🟢 Low | {total_low:,} |
| Corregibles (fix disponible) | {total_fixable:,} ({total_fixable/total_vulns*100:.1f}%) |
| CVSS máximo | {df['max_cvss'].max():.1f} |
"""
display(Markdown('### Resumen global de vulnerabilidades'))
display(Markdown(summary_md))
````

````python
# ── Distribución de severidad ─────────────────────────────────────────────────
SEVERITY_COLORS = {
    'Critical': '#d62728',
    'High':     '#ff7f0e',
    'Medium':   '#f7c948',
    'Low':      '#2ca02c',
    'Negligible':'#aec7e8',
    'Unknown':  '#c7c7c7',
}

sev_series = pd.Series({
    'Critical':   total_critical,
    'High':       total_high,
    'Medium':     total_medium,
    'Low':        total_low,
    'Negligible': df['negligible'].sum(),
    'Unknown':    df['unknown'].sum(),
})
sev_series = sev_series[sev_series > 0]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Barras
bar_colors = [SEVERITY_COLORS.get(s, '#999') for s in sev_series.index]
axes[0].bar(sev_series.index, sev_series.values, color=bar_colors, edgecolor='white', linewidth=0.5)
axes[0].set_title('Vulnerabilidades por severidad (total)', fontweight='bold')
axes[0].set_ylabel('Cantidad')
for i, v in enumerate(sev_series.values):
    axes[0].text(i, v + sev_series.max()*0.01, f'{v:,}', ha='center', fontweight='bold')

# Pie
axes[1].pie(sev_series, labels=sev_series.index,
            colors=bar_colors,
            autopct=lambda p: f'{p:.1f}%\n({int(round(p*total_vulns/100)):,})',
            startangle=90, pctdistance=0.75)
axes[1].set_title('Proporción de severidades', fontweight='bold')

plt.tight_layout()
plt.savefig(REPORTS / 'fig_severity_distribution.png', bbox_inches='tight')
plt.show()
````

````python
# ── Top 20 repos más vulnerables ─────────────────────────────────────────────
top20 = df.head(20).copy()

fig, ax = plt.subplots(figsize=(14, 8))
severities = ['critical', 'high', 'medium', 'low', 'negligible', 'unknown']
sev_labels  = ['Critical', 'High', 'Medium', 'Low', 'Negligible', 'Unknown']
colors      = [SEVERITY_COLORS[s] for s in sev_labels]

bottom = np.zeros(len(top20))
for col, label, color in zip(severities, sev_labels, colors):
    vals = top20[col].values
    if vals.sum() == 0:
        continue
    ax.barh(top20['repo'][::-1], vals[::-1], left=bottom[::-1],
            label=label, color=color, edgecolor='white', linewidth=0.3)
    bottom += vals  # acumular hacia adelante

ax.set_title('Top 20 repositorios con más vulnerabilidades\n(stacked por severidad)', fontweight='bold')
ax.set_xlabel('Número de vulnerabilidades')
ax.legend(loc='lower right')
plt.tight_layout()
plt.savefig(REPORTS / 'fig_top20_repos.png', bbox_inches='tight')
plt.show()
````

````python
# ── Distribución CVSS ────────────────────────────────────────────────────────
all_vulns = []
for r in results:
    for v in r.get('vulnerabilities', []):
        for cvss_entry in v.get('cvss', []):
            score = cvss_entry.get('metrics', {}).get('baseScore')
            if score is not None:
                all_vulns.append({
                    'repo': r['repo'],
                    'cve_id': v['id'],
                    'severity': v['severity'],
                    'cvss_score': float(score),
                    'fix_state': v.get('fix_state', 'unknown'),
                    'package_type': v.get('package_type', 'unknown'),
                })
                break  # un score por CVE

df_vulns = pd.DataFrame(all_vulns)
print(f'CVEs con score CVSS: {len(df_vulns):,}')

if not df_vulns.empty:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].hist(df_vulns['cvss_score'], bins=20, color='steelblue', edgecolor='white')
    axes[0].axvline(7.0, color='orange', linestyle='--', label='High (7.0)')
    axes[0].axvline(9.0, color='red',    linestyle='--', label='Critical (9.0)')
    axes[0].set_title('Distribución de scores CVSS', fontweight='bold')
    axes[0].set_xlabel('CVSS Base Score')
    axes[0].set_ylabel('Frecuencia')
    axes[0].legend()

    cvss_by_sev = df_vulns.groupby('severity')['cvss_score'].mean().sort_values(ascending=False)
    bar_c = [SEVERITY_COLORS.get(s, '#999') for s in cvss_by_sev.index]
    axes[1].bar(cvss_by_sev.index, cvss_by_sev.values, color=bar_c)
    axes[1].set_title('CVSS promedio por categoría de severidad', fontweight='bold')
    axes[1].set_ylabel('CVSS promedio')
    axes[1].set_ylim(0, 10)

    plt.tight_layout()
    plt.savefig(REPORTS / 'fig_cvss.png', bbox_inches='tight')
    plt.show()
````

````python
# ── Fixability: corregibles vs no corregibles ────────────────────────────────
if not df_vulns.empty:
    fix_counts = df_vulns['fix_state'].value_counts()
    fig, ax = plt.subplots(figsize=(7, 5))
    fix_colors = {'fixed': '#2ca02c', 'not-fixed': '#d62728', 'wont-fix': '#ff7f0e', 'unknown': '#c7c7c7'}
    bar_colors = [fix_colors.get(s, '#999') for s in fix_counts.index]
    ax.bar(fix_counts.index, fix_counts.values, color=bar_colors, edgecolor='white')
    ax.set_title('Estado de corrección de vulnerabilidades', fontweight='bold')
    ax.set_ylabel('Cantidad')
    for i, v in enumerate(fix_counts.values):
        ax.text(i, v + fix_counts.max()*0.01, f'{v:,}', ha='center')
    plt.tight_layout()
    plt.savefig(REPORTS / 'fig_fixability.png', bbox_inches='tight')
    plt.show()

    pct_fixable = (fix_counts.get('fixed', 0) / fix_counts.sum()) * 100
    print(f'Porcentaje de vulnerabilidades con fix disponible: {pct_fixable:.1f}%')
````

````python
# ── CVEs más frecuentes (Top 15) ─────────────────────────────────────────────
all_cve_list = []
for r in results:
    for v in r.get('vulnerabilities', []):
        all_cve_list.append({
            'cve_id': v['id'],
            'severity': v['severity'],
            'package': v.get('package', '?'),
        })

df_cves_all = pd.DataFrame(all_cve_list)
if not df_cves_all.empty:
    top_cves = df_cves_all['cve_id'].value_counts().head(15)
    # Obtener severidad de cada CVE
    cve_sev = df_cves_all.drop_duplicates('cve_id').set_index('cve_id')['severity']
    top_colors = [SEVERITY_COLORS.get(cve_sev.get(cve, 'Unknown'), '#999') for cve in top_cves.index]

    plt.figure(figsize=(13, 6))
    bars = plt.barh(top_cves.index[::-1], top_cves.values[::-1], color=top_colors[::-1])
    plt.title('Top 15 CVEs más frecuentes\n(cantidad de repos afectados)', fontweight='bold')
    plt.xlabel('Repositorios afectados')
    for bar in bars:
        w = bar.get_width()
        plt.text(w + 0.1, bar.get_y() + bar.get_height()/2, str(int(w)), va='center', fontsize=9)

    # Leyenda de severidad
    from matplotlib.patches import Patch
    legend_elements = [Patch(facecolor=SEVERITY_COLORS[s], label=s)
                       for s in ['Critical','High','Medium','Low'] if s in SEVERITY_COLORS]
    plt.legend(handles=legend_elements, loc='lower right')
    plt.tight_layout()
    plt.savefig(REPORTS / 'fig_top_cves.png', bbox_inches='tight')
    plt.show()
````

````python
# ── Heatmap: vulnerabilidades por repo y severidad ───────────────────────────
heat_cols = ['critical','high','medium','low']
heat_data = df.set_index('repo')[heat_cols].head(25)

if heat_data.shape[0] >= 3:
    plt.figure(figsize=(9, max(6, heat_data.shape[0] * 0.4)))
    sns.heatmap(
        heat_data,
        annot=True, fmt='d', cmap='YlOrRd',
        linewidths=0.4, linecolor='white',
        cbar_kws={'label': 'Cantidad de vulnerabilidades'},
    )
    plt.title('Heatmap: vulnerabilidades por repositorio y severidad\n(top 25)', fontweight='bold')
    plt.xticks(rotation=0)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.savefig(REPORTS / 'fig_heatmap.png', bbox_inches='tight')
    plt.show()
````

---
## 5. Correlaciones y hallazgos

````python
# ── Correlación: componentes vs vulnerabilidades ─────────────────────────────
df_merged = df_sbom_summary.merge(df[['repo','total_vulns','critical','high','max_cvss']], on='repo', how='inner')

if len(df_merged) > 3:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    axes[0].scatter(df_merged['total_components'], df_merged['total_vulns'],
                    alpha=0.7, color='steelblue', edgecolors='white', s=80)
    # Línea de tendencia
    z = np.polyfit(df_merged['total_components'], df_merged['total_vulns'], 1)
    p = np.poly1d(z)
    xp = np.linspace(df_merged['total_components'].min(), df_merged['total_components'].max(), 100)
    axes[0].plot(xp, p(xp), 'r--', alpha=0.8, label='Tendencia')
    axes[0].set_xlabel('Componentes en SBOM')
    axes[0].set_ylabel('Total vulnerabilidades')
    axes[0].set_title('Componentes vs Vulnerabilidades', fontweight='bold')
    axes[0].legend()

    corr = df_merged['total_components'].corr(df_merged['total_vulns'])
    axes[0].text(0.05, 0.95, f'r = {corr:.2f}', transform=axes[0].transAxes,
                 va='top', fontsize=11, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

    # Scatter CVSS máximo
    sc = axes[1].scatter(df_merged['total_components'], df_merged['max_cvss'],
                         c=df_merged['critical'], cmap='Reds', alpha=0.8,
                         s=90, edgecolors='grey', linewidths=0.5)
    plt.colorbar(sc, ax=axes[1], label='Vulns Critical')
    axes[1].set_xlabel('Componentes en SBOM')
    axes[1].set_ylabel('CVSS score máximo')
    axes[1].set_title('Componentes vs CVSS máximo\n(color = # críticas)', fontweight='bold')

    plt.tight_layout()
    plt.savefig(REPORTS / 'fig_correlations.png', bbox_inches='tight')
    plt.show()

    print(f'Correlación (componentes ~ vulnerabilidades): r = {corr:.3f}')
````

---
## 6. Tabla resumen final

````python
# ── Top repos de mayor riesgo ────────────────────────────────────────────────
df_risk = df.copy()
df_risk['risk_score'] = (
    df_risk['critical'] * 10 +
    df_risk['high']     *  5 +
    df_risk['medium']   *  2 +
    df_risk['low']      *  1
)
df_risk = df_risk.sort_values('risk_score', ascending=False)

display(Markdown('### Top 15 repositorios por puntuación de riesgo'))
display(Markdown('> Fórmula: `Critical×10 + High×5 + Medium×2 + Low×1`'))
display(
    df_risk[['repo','total_vulns','critical','high','medium','low','fixable','max_cvss','risk_score']]
    .head(15)
    .rename(columns={
        'repo':'Repositorio', 'total_vulns':'Total', 'critical':'Critical',
        'high':'High', 'medium':'Medium', 'low':'Low',
        'fixable':'Corregibles', 'max_cvss':'CVSS máx', 'risk_score':'Risk Score'
    })
    .style
    .background_gradient(subset=['Risk Score'], cmap='Reds')
    .background_gradient(subset=['Critical'], cmap='OrRd')
    .format({'CVSS máx': '{:.1f}', 'Risk Score': '{:,.0f}'})
)
````

````python
# ── Conclusiones ─────────────────────────────────────────────────────────────
total_repos_ok = len(df)
pct_critical_repos = (df['critical'] > 0).sum() / total_repos_ok * 100
pct_high_repos     = (df['high'] > 0).sum()     / total_repos_ok * 100
avg_vulns          = df['total_vulns'].mean()
fix_pct            = (total_fixable / total_vulns * 100) if total_vulns > 0 else 0
top_risky          = df_risk.iloc[0]['repo'] if len(df_risk) > 0 else 'N/A'

conclusions = f"""
## Conclusiones del análisis

### Organización analizada: `{ORG}`

1. **Cobertura**: Se analizaron **{total_repos_ok} repositorios activos** (último commit ≤30 días).

2. **Dependencias**: El total de componentes detectados en los SBOMs fue de **{df_sbom_summary['total_components'].sum():,}**,
   distribuidos principalmente entre los ecosistemas: {', '.join(eco_total.head(3).index.tolist())}.

3. **Vulnerabilidades**: Se encontraron **{total_vulns:,} vulnerabilidades** en total:
   - 🔴 Critical: {total_critical:,} ({total_critical/total_vulns*100:.1f}%)
   - 🟠 High:     {total_high:,} ({total_high/total_vulns*100:.1f}%)
   - 🟡 Medium:   {total_medium:,} ({total_medium/total_vulns*100:.1f}%)

4. **Exposición**: El **{pct_critical_repos:.0f}%** de los repositorios tiene al menos una vulnerabilidad crítica.
   El promedio por repo es de **{avg_vulns:.1f}** vulnerabilidades.

5. **Remediación**: El **{fix_pct:.1f}%** de las vulnerabilidades tiene un fix disponible,
   lo que representa una oportunidad concreta de reducir la superficie de ataque.

6. **Repo de mayor riesgo**: `{top_risky}` (Risk Score: {df_risk.iloc[0]['risk_score']:,.0f}).

### Recomendaciones
- Priorizar la actualización de dependencias en los repos con vulnerabilidades **Critical** y **High** con fix disponible.
- Integrar Syft + Grype en el CI/CD para detección continua.
- Establecer una política de SLAs: Critical → fix en 7 días, High → 30 días.
"""

display(Markdown(conclusions))
````
